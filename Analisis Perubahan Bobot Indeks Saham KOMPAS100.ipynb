```python
import pandas as pd
```

# Membaca dan Membersihkan Data

```python
file_path = '/content/4 Lamp Peng-00073-BEI POP - KOMPAS100 - Apr 2025 Minor.xlsx'
df = pd.read_excel(file_path, sheet_name='Lampiran 4')
```

```python
cleaned_df = df.iloc[7:, [2, 3, 4, 5, 6, 7, 8]]
cleaned_df.columns = [
    "Kode",
    "Rasio Free Float",
    "Jumlah Saham untuk Indeks (Pra Evaluasi)",
    "Jumlah Saham untuk Indeks (Pasca Evaluasi)",
    "Keterangan",
    "Bobot pada Indeks (Pra Evaluasi)",
    "Bobot pada Indeks (Pasca Evaluasi)"
]
```

```python
cleaned_df = cleaned_df.dropna().reset_index(drop=True)
print(cleaned_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Boxplot: Distribusi Rasio Free Float

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=cleaned_df, x='Kode', y='Rasio Free Float', hue='Keterangan')
plt.title('Scatterplot Rasio Free Float')
plt.show()
```

Pada boxplot distribusi rasio free float, persebaran data menggambarkan bagaimana rasio free float bervariasi di antara saham-saham KOMPAS100. Jika kotak cenderung miring atau terdapat titik-titik outlier, hal ini menunjukkan adanya beberapa saham dengan rasio yang sangat tinggi atau rendah dibandingkan mayoritas saham lain dalam indeks. Median yang tidak berada di tengah kotak menunjukkan bahwa distribusi rasio free float mungkin tidak simetris, mencerminkan adanya perbedaan signifikan antara saham dengan rasio tinggi dan rendah.

# Barplot: Bobot pada Indeks Sebelum dan Sesudah Evaluasi

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=cleaned_df, x='Kode', y='Bobot pada Indeks (Pra Evaluasi)', color='blue', alpha=0.6, label='Pra Evaluasi')
sns.barplot(data=cleaned_df, x='Kode', y='Bobot pada Indeks (Pasca Evaluasi)', color='orange', alpha=0.6, label='Pasca Evaluasi')
plt.title('Perbandingan Bobot pada Indeks (Pra vs Pasca Evaluasi)')
plt.xlabel('Kode Saham')
plt.ylabel('Bobot pada Indeks')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

Barplot yang menunjukkan perbandingan bobot pada indeks sebelum dan sesudah evaluasi menggambarkan perubahan pengaruh saham dalam indeks KOMPAS100. Jika bobot saham setelah evaluasi lebih rendah dibandingkan sebelumnya, berarti saham tersebut mengalami penurunan kontribusi dalam indeks, yang mungkin disebabkan oleh penurunan kapitalisasi pasar atau perubahan likuiditas. Sebaliknya, jika bobot meningkat, saham tersebut menjadi lebih dominan dalam indeks setelah evaluasi.

# Pie Chart: Distribusi Keterangan (Tetap/Naik/Turun/Baru)

```python
plt.figure(figsize=(6, 6))
cleaned_df['Keterangan'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['salmon', 'maroon', 'lightgrey', 'orange'])
plt.title('Distribusi Keterangan Status Saham')
plt.ylabel('')
plt.show()
```

Pada pie chart distribusi keterangan saham, bagian terbesar menunjukkan mayoritas saham yang tetap berada dalam indeks tanpa perubahan status. Jika ada porsi yang cukup besar untuk kategori "Naik" atau "Turun", itu mengindikasikan adanya perubahan signifikan dalam komposisi indeks. Adanya kategori "Baru" menandakan bahwa beberapa saham baru ditambahkan ke dalam indeks setelah proses evaluasi.

# Scatterplot: Perbandingan Jumlah Saham Sebelum dan Sesudah Evaluasi

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=cleaned_df, x='Jumlah Saham untuk Indeks (Pra Evaluasi)', y='Jumlah Saham untuk Indeks (Pasca Evaluasi)', hue='Keterangan')
plt.title('Scatterplot Jumlah Saham (Pra vs Pasca Evaluasi)')
plt.xlabel('Jumlah Saham Pra Evaluasi')
plt.ylabel('Jumlah Saham Pasca Evaluasi')
plt.show()
```

Scatterplot yang membandingkan jumlah saham sebelum dan sesudah evaluasi menunjukkan stabilitas atau perubahan dalam jumlah saham yang digunakan dalam indeks. Titik-titik yang berdekatan dengan garis diagonal menunjukkan konsistensi jumlah saham sebelum dan sesudah evaluasi. Jika ada titik yang jauh di atas atau di bawah garis tersebut, hal itu menandakan adanya perubahan signifikan dalam jumlah saham pada periode evaluasi. Warna-warna pada plot ini juga memberikan gambaran tentang status perubahan saham, apakah tetap, naik, turun, atau baru masuk ke dalam indeks.

# Histogram: Distribusi Bobot pada Indeks (Pra dan Pasca Evaluasi)

```python
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_df['Bobot pada Indeks (Pra Evaluasi)'], bins=20, color='blue', alpha=0.6, label='Pra Evaluasi')
sns.histplot(cleaned_df['Bobot pada Indeks (Pasca Evaluasi)'], bins=20, color='green', alpha=0.6, label='Pasca Evaluasi')
plt.title('Distribusi Bobot pada Indeks (Pra vs Pasca Evaluasi)')
plt.xlabel('Bobot pada Indeks')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

Histogram ini menggambarkan distribusi bobot saham pada indeks KOMPAS100 sebelum dan sesudah evaluasi. Jika histogram setelah evaluasi bergeser ke kanan, ini menandakan adanya peningkatan bobot untuk sebagian besar saham. Sebaliknya, pergeseran ke kiri menunjukkan penurunan bobot secara umum. Tumpang tindih antara kedua histogram menandakan adanya saham dengan bobot yang relatif stabil

# Line Plot: Tren Bobot pada Indeks untuk Saham Terpilih

```python
selected_stocks = ['BBCA', 'ASII', 'ANTM', 'ADRO', 'ACES']
subset = cleaned_df[cleaned_df['Kode'].isin(selected_stocks)]

plt.figure(figsize=(10, 6))
for kode in selected_stocks:
    stock_data = subset[subset['Kode'] == kode]
    plt.plot(['Pra Evaluasi', 'Pasca Evaluasi'],
             [stock_data['Bobot pada Indeks (Pra Evaluasi)'].values[0],
              stock_data['Bobot pada Indeks (Pasca Evaluasi)'].values[0]],
             marker='o', label=kode)

plt.title('Tren Bobot pada Indeks Saham Terpilih')
plt.xlabel('Evaluasi')
plt.ylabel('Bobot pada Indeks')
plt.legend()
plt.show()
```

Plot garis ini menggambarkan perubahan bobot pada indeks untuk beberapa saham terpilih. Garis yang menurun menunjukkan penurunan bobot setelah evaluasi, sedangkan garis yang naik menandakan peningkatan. Saham dengan garis mendatar berarti bobotnya tetap stabil. Plot ini membantu memvisualisasikan bagaimana evaluasi memengaruhi bobot pada indeks saham utama secara lebih mendetail.
